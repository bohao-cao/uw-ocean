```python
import pandas as pd
post_df = pd.read_parquet("../concat_feature.parquet")
```

```python
post_df=post_df.dropna()
```

```python
import re
post_df['num_hashtags'] = post_df['caption'].map(lambda r:len(re.findall("#", r)))
```

```python
post_df..value_counts()
```

```python
def get_hashtags(caption):
    result = []
    caption_arr = caption.split(" ")
    for str in caption_arr:
        if str.startswith("#"):
            result.append(str)
    return result

post_df['hashtags'] = post_df['caption'].apply(get_hashtags)
```

```python
hashtags_by_post = post_df[post_df['num_hashtags']>0].hashtags.tolist()
from itertools import chain
def flatten_chain(hashtags_2d):
    return list(chain.from_iterable(hashtags_2d))

hashtags_1d = flatten_chain(hashtags_by_post)
```

```python
from collections import Counter
hashtags_counter= Counter(hashtags_1d)
```
